## Your Task

Your task is to create a bar graph of the 25 most common words in Shakespeare's Macbeth.  


A common Python programming pattern to count objects, produce histograms, or update statistics is to make calls to a dictionary as you iterate through a list. For example, given a list of words, you can create a dictionary to store counts and then iterate through the list of words, checking how many times each word has appeared using your dictionary, and updating the dictionary count now that you've seen that word again. The `.get()` method of dictionary is very useful in doing this. Read the docstring for the `.get()` method and use it along with the pseudocode below to create a bar graph of the 25 most common words from the transcript of Macbeth which has been loaded into the variable 'macbeth'. Be sure to include a title and appropriate labels for your graph.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

#Slice transcript to abridge intro header material into macbeth_abr
macbeth_abr = macbeth [16648: ]

#Remove new lines into macbeth_cln
macbeth_cln = macbeth_abr .replace( '\n' , ' ' ) .replace( '\r' , '' )

#split words into macbeth_rawlist
macbeth_rawlist = macbeth_cln.split()

#Clean punctuation into macbeth_cleanlist
macbeth_cleanlist = []
for word in macbeth_rawlist:
    cleanword = word .strip() .strip( "," ) .strip( "(" ) .strip( ")" ).strip( "." ).strip( "?" ).strip( "!" ).strip( ":" ).strip( ";" )
    macbeth_cleanlist .append( cleanword )

#creating blank dict of unique wordsd
macbeth_unique_list = list( set( macbeth_cleanlist ) )
macbeth_countsdict = dict .fromkeys( macbeth_unique_list , 0 )

#iterating counts of words
for word in macbeth_cleanlist:
    macbeth_countsdict[ word ] += 1

def sortfunc(x):
    return macbeth_countsdict [ x ]
    
macbeth_sorted_uniques = sorted( macbeth_countsdict .keys() , key = sortfunc, reverse = True)
macbeth_sorted_counts = sorted( macbeth_countsdict .values() , reverse = True)
#print(macbeth_sorted_uniques)
#print(macbeth_sorted_counts)


# Create Bar Graph of top 200 most common words big enough to show all the xlabel words
plt.figure( figsize = ( 50, 10), dpi = 200, facecolor = 'w', edgecolor = 'k')
plt.bar( macbeth_sorted_uniques[ :200],macbeth_sorted_counts[ :200])

# Include descriptive titles and labels
plt.xlabel( 'Unique Words' )
plt.ylabel( 'Count of Words' )
plt.title( 'Word Frequency in Macbeth' )
plt.xticks( rotation = 90 )

#Suppresses some of the unnecessary output of plt.bar or plt.xticks
plt.show()


```
